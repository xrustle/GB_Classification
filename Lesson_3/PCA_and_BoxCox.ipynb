```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import random
```

## Логарифмируем признак

```python
p1x = np.random.randn((1000)) + 10

p2x = np.random.randn((1000)) + 13

plt.figure(figsize=(10, 5))

sns.kdeplot(p1x, shade=True, label='No default', color='g')
sns.kdeplot(p2x, shade=True, label='Default', color='r')
plt.show()
```

На вход не нормальное распределение

```python
p1xe = np.exp(p1x)
p2xe = np.exp(p2x)

plt.figure(figsize=(10, 5))
sns.kdeplot(p1xe, shade=True, label='No default', color='g')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(p2xe, shade=True, label='Default', color='r')
plt.show()
```

## Бокс-Кокс

```python
from scipy.stats import boxcox

data1, lmbda = boxcox(p1xe)
print(lmbda)

plt.figure(figsize=(10, 5))
sns.kdeplot(data1, shade=True, label='Default', color='r')
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA

N = 1000
N2 = 500

X1 = np.linspace(0, 100, N) + np.random.randn(N)
X2 = np.linspace(0, 100, N) + np.random.randn(N)

X = np.vstack([X1, X2])
X.shape
```

```python
pca = PCA(n_components=2)
pca.fit(X.T)
Xp = pca.transform(X.T)
Xp.shape
```

```python
plt.plot(X1[:N2], X2[:N2], 'or')
plt.plot(X1[N2:], X2[N2:], 'og')
plt.title('before PCA')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
plt.plot(Xp[:N2, 1], Xp[:N2, 0], 'ob')
plt.plot(Xp[N2:, 1], Xp[N2:, 0], 'og')
plt.title('after PCA')
plt.xlabel('comp1')
plt.ylabel('comp0')
plt.show()
```
